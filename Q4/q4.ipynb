```python
import csv
import matplotlib.pyplot as plt

if __name__ == "__main__":
    f = open('seoul_subway.csv', encoding='ANSI')
    data = csv.reader(f)
    
    header = next(data)
    next(data)
    
    in_dic = {}
    out_dic = {}
    inout_dic = {}
    
    in_list = []
    out_list = []
    inout_list = []
    
    max_p = [[0]*30,[0]*30,[0]*30]
    max_st = [[0]*30,[0]*30,[0]*30]
    
    for line in data:
        
        tin = int(line[10])+int(line[12])
        tout = int(line[11])+int(line[13])
        tinout = tin + tout
        
        in_dic[tin] = line[3]
        out_dic[tout] = line[3]
        inout_dic[tinout] = line[3]
        
    f.close()
        
    in_list = list(in_dic.keys())    
    out_list = list(out_dic.keys())
    inout_list = list(inout_dic.keys())
    
    in_list.sort(reverse=True)
    out_list.sort(reverse=True)
    inout_list.sort(reverse=True)
    
    
    
    for i in range(30) : 
        key = int(in_list[i])
        max_p[0][i] = key
        max_st[0][i] = in_dic.get(key)
        
        key = int(out_list[i])
        max_p[1][i] = key
        max_st[1][i] = out_dic.get(key)
        
        key = int(inout_list[i])
        max_p[2][i] = key
        max_st[2][i] = inout_dic.get(key)
    
    
    plt.rc('font',family = 'Malgun Gothic')
    plt.rcParams['axes.unicode_minus'] = False
    
    plt.figure(figsize=(50,5))
    plt.subplot(1,3,1)
    plt.bar(max_st[0],max_p[0])
    plt.xticks(max_st[0],rotation = 90)
    plt.xlabel(' ')
    plt.ylabel('')
    plt.title(" ")
    plt.show()
    
    plt.figure(figsize=(50,5))
    plt.subplot(1,3,2)
    plt.bar(max_st[1],max_p[1])
    plt.xticks(max_st[1],rotation = 90)
    plt.xlabel(' ')
    plt.ylabel('')
    plt.title(" ")
    plt.show()
    
    plt.figure(figsize=(50,5))
    plt.subplot(1,3,3)
    plt.bar(max_st[2],max_p[2])
    plt.xticks(max_st[2],rotation = 90)
    plt.xlabel(' ')
    plt.ylabel('')
    plt.title(" ")
    plt.show()
```
