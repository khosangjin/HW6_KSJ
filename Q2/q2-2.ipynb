```python
import matplotlib.pyplot as plt
import csv

if __name__ == ("__main__"):
    
    f = open('dice.csv', encoding='ANSI')
    data = csv.reader(f)
    line = list(data)
    
    result = [[],[],[],[]]       
    frequency = [[0]*6,[0]*6,[0]*6,[0]*6]
    
    k=0
    for i in line:
        for j in i:
            frequency[k][int(j)-1] += 1
        k += 1
        
    for i in frequency:
        print(i)
        
    plt.rc('font',family = 'Malgun Gothic')
    plt.rcParams['axes.unicode_minus'] = False
    
    for i in range(len(line)):
        plt.figure(figsize=(50,5))
        plt.subplot(1,4,i+1)
        result[i] = [int(x) for x in line[i]]
        plt.hist(result[i],bins=11)
        plt.xlabel(' ')
        plt.ylabel('')
        plt.title("{0:d} ".format(len(result[i])))
        plt.grid(True,axis='y')
        plt.show()
  
```
