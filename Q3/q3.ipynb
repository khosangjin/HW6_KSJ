```python
import csv
import matplotlib.pyplot as plt

if __name__ == "__main__":
    f = open('Jeju_population.csv', encoding='ANSI')
    data = csv.reader(f)
    header = next(data)

    line = next(data)
    #print(line)
    year = []
    male = []
    female = []
    
    for i in range(10):
        year.append(2013+i)
        male.append(int(line[3+4*i]))
        female.append(int(line[4+4*i]))
        print("Year = {0:d}, male = {1:d}, female = {2:d}, diff = {3:d}".format(year[i],male[i],female[i],male[i]-female[i]))

    f.close()    

    plt.plot(year,male,color='blue',label='male')
    plt.plot(year,female,color='red',label='female')
    plt.xticks(range(2013, 2023, 1), year)
    plt.xlabel('year')
    plt.ylabel('population')
    plt.legend()
    plt.grid(True,axis='x')
    plt.grid(True,axis='y')
    plt.show
```
