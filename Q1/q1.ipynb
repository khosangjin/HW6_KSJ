```python
import csv
import matplotlib.pyplot as plt

if __name__ == "__main__":
    month = []
    whole = []
    seoul = []
    daejeon = []
    busan = []
    jeju = []
    
    for i in range(12):
        month.append(1+i)
    
    #전국
    f = open('whole.csv', encoding='ANSI')
    data = csv.reader(f)
    next(data)
    next(data)
    next(data)
    next(data)
    next(data)
    next(data)
    next(data)
    header = next(data)
    
    for line in data:
        whole.append(float(line[2]))
        
    #서울
    f = open('seoul.csv', encoding='ANSI')
    data = csv.reader(f)
    next(data)
    next(data)
    next(data)
    next(data)
    next(data)
    next(data)
    next(data)
    header = next(data)
    
    for line in data:
        seoul.append(float(line[2]))
    
    #대전
    f = open('daejeon.csv', encoding='ANSI')
    data = csv.reader(f)
    next(data)
    next(data)
    next(data)
    next(data)
    next(data)
    next(data)
    next(data)
    header = next(data)
    
    for line in data:
        daejeon.append(float(line[2]))
    
    #부산
    f = open('busan.csv', encoding='ANSI')
    data = csv.reader(f)
    next(data)
    next(data)
    next(data)
    next(data)
    next(data)
    next(data)
    next(data)
    header = next(data)
    
    for line in data:
        busan.append(float(line[2]))
        
    #제주
    f = open('jeju.csv', encoding='ANSI')
    data = csv.reader(f)
    next(data)
    next(data)
    next(data)
    next(data)
    next(data)
    next(data)
    next(data)
    header = next(data)
    
    for line in data:
        jeju.append(float(line[2]))
        
    print('전국',whole)
    print('서울',seoul)
    print('대전',daejeon)
    print('부산',busan)
    print('제주',jeju)
    
    #차트 크기
    plt.figure(figsize=(20,10))
    
    plt.bar(range(1,73,6),whole,color='black',label='whole')
    plt.bar(range(2,74,6),seoul,color='red',label='seoul')
    plt.bar(range(3,75,6),daejeon,color='orange',label='daejeon')
    plt.bar(range(4,76,6),busan,color='green',label='busan')
    plt.bar(range(5,77,6),jeju,color='blue',label='jeju')
    plt.xticks(range(1, 73, 6), month)
    plt.yticks(range(-5, 32, 2))
    plt.legend()
    plt.xlabel('month')
    plt.ylabel('average temperature')
    plt.grid(True,axis='y')
    plt.show()
```
